## Clean Timbre 

```python
timbre_features = ['timbre_all_1', 'timbre_all_2',
       'timbre_all_3', 'timbre_all_4', 'timbre_all_5', 'timbre_all_6',
       'timbre_all_7', 'timbre_all_8', 'timbre_all_9', 'timbre_all_10',
       'timbre_all_11', 'timbre_all_12']
df_timbre = df_songs[timbre_features]
df_timbre.head()
```

```python
# Check type of array:
type(df_timbre.iloc[0,0])
```

```python
# Function for converting TimbreString into List
def timbreStringToList(timbreString):
    timbreString = timbreString[:-1]
    timbreString = timbreString[1:]
    timbreList = [float(s) for s in timbreString.split(',')]
    return timbreList
```

```python
# Convert strings to lists
for i in range(df_timbre.shape[1]):
    timbre_df = df_timbre.iloc[:,i]
    for j in range(timbre_df.shape[0]):
        timbre_df[j] = timbreStringToList(timbre_df[j])
    df_timbre.iloc[:,i] = timbre_df
```

```python
# Test if it's a list
type(df_timbre.iloc[2,1])
```

```python
df_timbre.shape
```

```python
# Create timbre features
for i in range(df_timbre.shape[1]):
    timbre_df = df_timbre.iloc[:,i]
    timbre_mean = np.zeros(timbre_df.shape[0])
    timbre_median = np.zeros(timbre_df.shape[0])
    timbre_std = np.zeros(timbre_df.shape[0])
    timbre_min = np.zeros(timbre_df.shape[0])
    timbre_max = np.zeros(timbre_df.shape[0])
    timbre_range = np.zeros(timbre_df.shape[0])
    timbre_80Percentile = np.zeros(timbre_df.shape[0])
    for j in range(timbre_df.shape[0]):
        timbre_mean[j] = np.mean(timbre_df[j])
        timbre_median[j] = np.median(timbre_df[j])
        timbre_std[j] = np.std(timbre_df[j])
        timbre_min[j] = np.min(timbre_df[j])
        timbre_max[j] = np.max(timbre_df[j])
        timbre_range[j] = timbre_max[j]- timbre_min[j]
        timbre_80Percentile[j] = np.percentile(timbre_df[j], 80)
        df_timbre['mean_timbre{}'.format(i+1)] = timbre_mean
        df_timbre['median_timbre{}'.format(i+1)] = timbre_median
        df_timbre['std_timbre{}'.format(i+1)] = timbre_std
        df_timbre['min_timbre{}'.format(i+1)] = timbre_min
        df_timbre['max_timbre{}'.format(i+1)] = timbre_max
        df_timbre['range_timbre{}'.format(i+1)] = timbre_range
        df_timbre['80Percentile_timbre{}'.format(i+1)] = timbre_80Percentile
        if((j+1)%10 == 0):
            print('Finished {} rows within column {}'.format(j+1, i))
    print('Finished Column {}'.format(i))
```
