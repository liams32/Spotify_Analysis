# ML Models on Timbre Data

## Loading more or less clean data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_timbre = pd.read_csv('TimbreAll_withFeatures.csv')
df_timbre.head()
```

```python
df_timbre.columns
```

```python
# Drop columns
df_timbre = df_timbre.drop(['Unnamed: 0', 'Unnamed: 0.1', 'timbre_all_1', 'timbre_all_2',
       'timbre_all_3', 'timbre_all_4', 'timbre_all_5', 'timbre_all_6',
       'timbre_all_7', 'timbre_all_8', 'timbre_all_9', 'timbre_all_10',
       'timbre_all_11', 'timbre_all_12'], axis=1)
df_timbre.head()
```

```python
df_global = pd.read_csv('cleanGlobalPrediction1.csv',encoding='latin-1')
df_global.head()
```

```python
# Define dataframe
df_raw = pd.concat([df_global, df_timbre] , axis = 1)
df_raw.head()
```

```python
df_raw.columns
```

```python
df = df_raw.drop(['Unnamed: 0','analysis_url', 'track_href', 'uri'], axis =1)
print(df.shape)
df.head()
```

```python
# Devide popularity into quantile bins
#df['popularity'] = pd.qcut(df['popularity'], 3, labels= False)
#df['popularity']
```

```python
# Drop middle quantile to create gap
df_reduced = df[df.popularity!=1]
print('Reduced data set to:')
print(df.shape[0])
```

```python
features = list(df.columns)
features.remove('global')
features.remove('global_100')
features.remove('global_top')
features.remove('popularity')

features.remove('song')
features.remove('artist')
print('Those are our {} features'.format(len(features)))
print(features)
```

## Machine Learning

### Define Input/Output

```python
# Clean dataframe
df_reduced.head()
```

```python
# Input
X= df_reduced[features]
X.head()
print(X.shape)
```

```python
# Normalize Input
from sklearn.preprocessing import normalize
X_norm = normalize(X)
```

```python
# Output
Y = df_reduced['popularity']
print(Y.shape)
Y.head()
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.15, random_state=100)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
def printConfusionMatrix (y_true, y_pred):
    cf=pd.DataFrame(confusion_matrix(y_true, y_pred))
    print(cf)
```

### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
print('Logistic Regression on popularity: ')
logRegModel = LogisticRegression()
logRegModel.fit(x_train, y_train)
# Accuracy
print('Train Accuracy:')
print(logRegModel.score(x_train, y_train))
print('Test Accuracy:')
print(logRegModel.score(x_test, y_test))
print('Confusion Matrix')
printConfusionMatrix(y_test, logRegModel.predict(x_test))
```

### 2. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=190)
random_forest.fit(x_train, y_train)
acc_rf1 = random_forest.score(x_train, y_train)
acc_rf2 = random_forest.score(x_test, y_test)
print("Random Forest: ")
print("Training: ", acc_rf1)
print("Test: ", acc_rf2)
print('Confusion Matrix')
printConfusionMatrix(y_test, random_forest.predict(x_test))
```

```python
values = random_forest.feature_importances_
```

```python
dict_featureImportance = dict(zip(features,values))
dict_featureImportance
```
