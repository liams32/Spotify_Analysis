# Cleaning raw positions

## Import data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_rawPos = pd.read_csv('scraped_positions.csv')
df_rawPos.head()
```

```python
features_rawPos = list(df_rawPos.columns)
startIndex = features_rawPos.index('pt_position')
features_pos = features_rawPos[startIndex:]
print(features_pos)
```

```python
df_positions = df_rawPos[features_pos]
df_positions.head()
```

```python
# Function for converting TimbreString into List
def timbreStringToList(timbreString):
    timbreString = timbreString[:-1]
    timbreString = timbreString[1:]
    if len(timbreString)!=0:
        timbreList = [float(s) for s in timbreString.split(',')]
    else:
        timbreList = np.nan
    return timbreList
```

```python
df_positions.shape
```

```python
for i in range(df_positions.shape[1]):
    positions_df = df_positions.iloc[:,i]
    for j in range(df_positions.shape[0]):
        positions_df[j] = timbreStringToList(positions_df[j])
    df_positions.iloc[:,i] = positions_df
```

```python
df_positions.head()
```

```python
# Get averages
print('Getting started')
for i in range(df_positions.shape[1]):
    positions_df = df_positions.iloc[:,i]
    for j in range(df_positions.shape[0]):
        positions_df[j] = np.mean(positions_df[j])
        if j%100==0:
            print('Finished another 100')
    print('Finished column'.format(i))
```
