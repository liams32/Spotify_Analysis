### There are 19422 unique uris, so expect this to take about 5 hours to finish?

```python
genre_dict = {}
```

```python
%%time
for i in range(0, len(dat)):
    curr_uri = dat['uri'][i]
    if curr_uri not in genre_dict.keys():
        possible_genre = sp.artist(sp.track(dat['uri'][i])['artists'][0]['uri'])['genres']
        if possible_genre == []:
            genre_dict[curr_uri] = 'Unknown'
        else:
            genre_dict[curr_uri] = possible_genre[0]
    if i % 100 == 0:
        print(len(dat) - i, 'remaining')
```
