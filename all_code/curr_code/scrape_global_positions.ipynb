```python
import time #to track time
total_entries = dat.shape[0] * len(region_names) #to track if network broken in middle of loop
original_time = time.time() #to track time of loop

for name in region_names:
    new_col = name + '_position' #each country now has column to hold positions: column called (region)_position
    dat[new_col] = ''
    for song_index in range(0, len(dat)): #dat is our old df
        start_time = time.time() #to track time of loop
        #we use region == i to subset proper region and track name == track name[song_index] to subset song name
        song_positions = world[(world['Region'] == name) & (world['Track Name'] == world['Track Name'][song_index])]['Position']
        new_set = set(song_positions)
        new_entry = list(new_set)
        dat[new_col][song_index] = new_entry #append new column to dataframe
        
        total_entries -= 1
        original_time += time.time()
        
        print(total_entries, 'left to go', time.time() - start_time, ': time elapsed on this loop')
        
print(original_time, 'total time elapsed')
```
